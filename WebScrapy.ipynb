```python
import requests 
from bs4 import BeautifulSoup 
URL = "https://nips.cc/Conferences/2019/Schedule"
r = requests.get(URL) 
reponse = requests.get(URL,timeout=5)
content = BeautifulSoup(reponse.content,"html.parser")
```
